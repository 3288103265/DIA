```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('test_imgs/haze1.jpg')
```

```python
plt.imshow(img[:,:,::-1])
```

```python
height, width, _ = img.shape
```

```python
dark = np.zeros((height, width))
window_size = 15
padding = int((window_size-1)/2)
padding_img = 255*np.ones((height+window_size-1, width+window_size-1, 3), dtype=np.int8)
padding_img[padding:padding+height,padding:padding+width,:]=img
# padding_img = cv2.copyMakeBorder(img, padding, padding, padding, padding, cv2.BORDER_CONSTANT, value=255
plt.imshow(padding_img[:,:,::-1])
```

```python
for h in range(height):
    for w in range(width):
        dark[h][w]=padding_img[h:h+window_size,w:w+window_size,:].min()
```

```python
plt.imshow(dark,cmap='gray')
```
